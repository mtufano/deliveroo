# GET URLs

Note: the results can vary based on time at location - 
the keyword after https://deliveroo.co.uk/restaurants/london/ determines where the order should be delivered

```python
import requests
from bs4 import BeautifulSoup

postcodes = ['E', 'EC', 'N', 'NW','SE', 'SW', 'W', 'WC',]
#postcode = 'W1'
urls = []

for postcode in postcodes:
    url = f'https://deliveroo.co.uk/restaurants/london/mayfair/?postcode={postcode}&collection=all-restaurants'
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('/menu/'):
            modified_href = 'https://deliveroo.co.uk' + href + ','
            urls.append(modified_href)

link_count = len(urls)
print("Number of links:", link_count)
print("Modified URLs:")
for url in urls:
    print(url)
```

```python
import requests
from bs4 import BeautifulSoup

postcodes = ['E', 'EC', 'N', 'NW', 'SE', 'SW', 'W', 'WC']
urls = []

for postcode in postcodes:
    url = f'https://deliveroo.co.uk/restaurants/london/mayfair/?postcode={postcode}&collection=all-restaurants'
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('/menu/'):
            modified_href = 'https://deliveroo.co.uk' + href + ','
            urls.append(modified_href)

link_count = len(urls)
print("Number of links:", link_count)

# Save the URLs in a text file
filename = "urls.txt"
with open(filename, "w") as file:
    file.write("\n".join(urls))

print("URLs saved to", filename)
```

```python
filename = "urls.txt"
output_filename = "urls_london_mayfair_all_postcodes.txt"

with open(filename, "r") as file:
    content = file.read()

# Add quotes before "https" and before the comma
modified_content = content.replace("https", "\"https").replace(",", "\",")

with open(output_filename, "w") as file:
    file.write(modified_content)

print("Modified URLs saved to", output_filename)
```
