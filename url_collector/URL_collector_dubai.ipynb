```python
import requests
from bs4 import BeautifulSoup

postcodes = ['E', 'EC', 'N', 'NW','SE', 'SW', 'W', 'WC',]
#postcode = 'W1'
urls = []

for postcode in postcodes:
    url = f'https://deliveroo.ae/restaurants/dubai/deira/?geohash=thrrg50szgsf&collection=restaurants'
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('/menu/'):
            modified_href = 'https://deliveroo.ae' + href + ','
            urls.append(modified_href)

link_count = len(urls)
print("Number of links:", link_count)
print("Modified URLs:")
for url in urls:
    print(url)
```
